```python
#Do the distributions for different numerical variables look like a normal distribution

from scipy.stats import norm 

numerical['Customer Lifetime Value'].mean()


np.std(numerical['Customer Lifetime Value'])

#Normaldistribution for customer lifetie value 

normal = norm(numerical['Customer Lifetime Value'].mean(),np.std(numerical['Customer Lifetime Value']))
x=np.linspace(numerical['Customer Lifetime Value'].max(),-numerical['Customer Lifetime Value'].max(),100)
plt.plot(x,normal.pdf(x))
     
```

```python
#Normaldistribution for customer lifetie value Monthly Premium Auto

normal = norm(numerical['Monthly Premium Auto'].mean(),np.std(numerical['Monthly Premium Auto']))
x=np.linspace(numerical['Monthly Premium Auto'].max(),-numerical['Monthly Premium Auto'].max(),100)
plt.plot(x,normal.pdf(x))
```

```python
#Normaldistribution for NUmber of Policies
normal = norm(numerical['Number of Policies'].mean(),np.std(numerical['Number of Policies']))
x=np.linspace(numerical['Number of Policies'].max(),-numerical['Number of Policies'].max(),100)
plt.plot(x,normal.pdf(x))
```
