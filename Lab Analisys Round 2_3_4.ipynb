```python
#Elimino la columna vehicle type por alto número de valores nulos
dfm=dfm.drop('vehicle_type', axis=1)
```

```python
#Columnas state, response, vechicle class, vehicle size pueden rellenarse con la moda al ser variables categoricas
mode_state=dfm['state'].mode()
dfm['state']=dfm['state'].fillna(mode_state[0])
```

```python
dfm
```

```python
mode_response=dfm['response'].mode()
dfm['response']=dfm['response'].fillna(mode_response[0])

mode_vehicle_class=dfm['vehicle_class'].mode()
dfm['vehicle_class']=dfm['vehicle_class'].fillna(mode_vehicle_class[0])

mode_vehicle_size=dfm['vehicle_size'].mode()
dfm['vehicle_size']=dfm['vehicle_size'].fillna(mode_vehicle_size[0])
```

```python
#Columnas months since last claim y number of open complaints con la media 
mslc_mean= dfm['months_since_last_claim'].mean()
dfm['months_since_last_claim']=dfm['months_since_last_claim'].fillna(mslc_mean)

noc_mean= dfm['number_of_open_complaints'].mean()
dfm['number_of_open_complaints']=dfm['number_of_open_complaints'].fillna(noc_mean)
```

```python
#Do the distributions for different numerical variables look like a normal distribution

from scipy.stats import norm 

numerical['customer_lifetime_value'].mean()


np.std(numerical['customer_lifetime_value'])

#Normaldistribution for customer lifetie value 

normal = norm(numerical['customer_lifetime_value'].mean(),np.std(numerical['customer_lifetime_value']))
x=np.linspace(numerical['customer_lifetime_value'].max(),-numerical['customer_lifetime_value'].max(),100)
plt.plot(x,normal.pdf(x))
```

```python
#normal distribution for monthly premium auto 

normal = norm(numerical['monthly_premium_auto'].mean(),np.std(numerical['monthly_premium_auto']))
x=np.linspace(numerical['monthly_premium_auto'].max(),-numerical['monthly_premium_auto'].max(),100)
plt.plot(x,normal.pdf(x))
```

```python
#Normaldistribution for NUmber of Policies
normal = norm(numerical['number_of_policies'].mean(),np.std(numerical['number_of_policies']))
x=np.linspace(numerical['number_of_policies'].max(),-numerical['number_of_policies'].max(),100)
plt.plot(x,normal.pdf(x))
```

```python

#normal distribution for number of policies with mean 0 
normal = norm(0,np.std(numerical['number_of_policies']))
x=np.linspace(numerical['number_of_policies'].max(),-numerical['number_of_policies'].max(),100)
plt.plot(x,normal.pdf(x))
     
```
