```python
for x in ['customer_lifetime_value']:
    q75,q25 = np.percentile(data_num.loc[:,x],[75,25])
    intr_qr = q75-q25
 
    max = q75+(1.5*intr_qr)
    min = q25-(1.5*intr_qr)
 
    data_num.loc[data_num[x] < min,x] = np.nan
    data_num.loc[data_num[x] > max,x] = np.nan
```
