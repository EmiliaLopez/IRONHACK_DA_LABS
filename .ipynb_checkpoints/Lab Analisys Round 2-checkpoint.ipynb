```python
##Check and deal with NaN values.
dfm.isnull().sum()
```

```python
for i in dfm.columns:
  nan_values=sum(dfm[i].isnull())
  #si no tienen valores NaN no se mostrarán
  if nan_values > 0:
    print('En la columna: ',i,'hay ',nan_values,'valores nulos')
```

```python
#Elimino la columna vehicle type por alto número de valores nulos

dfm=dfm.drop('vehicle type', axis=1)
dfm
```

```python
#Columnas state, response, vechicle class, vehicle size pueden rellenarse con la moda 

import numpy as np

mode_state=dfm['state'].mode()
dfm['state']=dfm['state'].fillna(mode_state)

```

```python
#Columnas months since last claim y number of open complaints con la media
mslc_mean= dfm['months since last claim'].mean()
dfm['months since last claim']=dfm['months since last claim'].fillna(mslc_mean)

noc_mean= dfm['number of open complaints'].mean()
dfm['number of open complaints ']=dfm['number of open complaints'].fillna(noc_mean)
```

```python
dfm.isnull().sum()
```

```python
#Datetime format - Extract the months from the dataset and store in a separate column. Then filter the data to show only the information for the first quarter , ie. January, February and March. Hint: If data from March does not exist, consider only January and February.
```

```python
dfm['effective to date'] = pd.to_datetime(dfm['effective to date'])
dfm.dtypes
```

```python
dfm['month']= dfm['effective to date'].dt.month
dfm
```

```python
#BONUS: Put all the previously mentioned data transformations into a function.
def clean_data(dfm):
     dfm.columns=[e.lower().replace(' ', '_') for e in dfm.columns]
     dfm=dfm.drop(columns=['unnamed:_0', 'vehicle_type'])
     dfm=dfm.dropna()    
     dfm['effective_to_date']=pd.to_datetime(dfm['effective_to_date'], errors='coerce')
     return dfm
```
