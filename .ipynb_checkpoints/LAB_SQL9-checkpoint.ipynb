```python
from sqlalchemy import create_engine
import pandas as pd
import getpass
import pymysql
```

```python
password = getpass.getpass()
```

```python
connection_string = 'mysql+pymysql://root:2710@127.0.0.1/sakila'
engine = create_engine(connection_string)
```

```python
query='SELECT customer_id, COUNT(*) as num_rentals \
FROM rentals_may \
GROUP BY customer_id;'
data_may=pd.read_sql_query(query, engine)
```

```python
data_may
```

```python
#Check the number of rentals for each customer for June
```

```python
query='SELECT customer_id, COUNT(*) as num_rentals \
FROM rentals_june \
GROUP BY customer_id;'
data_june=pd.read_sql_query(query, engine)
```

```python
data_june
```

```python
#Write a function that checks if customer borrowed more or less films in the month of June as compared to May.
```

```python
result = pd.merge(data_may, data_june, how='inner', on="customer_id",
                 suffixes=('_may', '_june'))
result
```
