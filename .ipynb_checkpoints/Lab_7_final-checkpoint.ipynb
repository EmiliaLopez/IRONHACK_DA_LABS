```python
data = pd.read_csv('marketing_customer_analysis.csv')
```

```python
#Estadarizamos headers
def clean_headers(x):
  x.columns = [i.lower().replace(' ','_') for i in x.columns]
  x['effective_to_date'] = pd.to_datetime(x['effective_to_date'],errors='coerce')
  return x
```

```python
#Vemos que number of open complaints nos da mas de un 75% de informacion vacia, por lo tanto se borrarÃ¡ y las demas columnas se llenaran con su promedio 
def data_clean(x):
  x = x.drop('number_of_open_complaints',axis=1)
  for i in x.columns:
    x[i] = x[i].replace(0,stat.mean(x[i]))
  return x 
```
