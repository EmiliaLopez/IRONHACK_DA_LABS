```python
#X-y split.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('marketing_customer_analysis.csv')
```

```python
#Tratamos datos

def clean_data(dfm):
     dfm.columns=[e.lower().replace(' ', '_') for e in dfm.columns]
     dfm=dfm.drop(columns=['unnamed:_0', 'vehicle_type'])
     dfm=dfm.dropna()    
     dfm['effective_to_date']=pd.to_datetime(dfm['effective_to_date'], errors='coerce')
     return dfm
```

```python
df= clean_data(df)
```

```python
df
```

```python
#X-y split.
#total_claim_amount tiene mayor valor de multicolinealidad por lo que será la variable dependiente y pasamos a elimanarla y asignarle Y

X=df.drop('total_claim_amount', axis=1)
y=df.total_claim_amount
```

```python
#Normalize (numerical).

from scipy.spatial import transform
from sklearn.preprocessing import Normalizer
from sklearn.preprocessing import StandardScaler 
```

```python
#Se divide numérica y categorica 
X_num = X.select_dtypes(include=np.number)
X_cat = X.select_dtypes(include=np.object)
```

```python
X_num.describe()
```

```python
#Normalizacion estandar 

#En la función normalize(X) se realiza la normalización estandar de la variable X, donde:

#X.mean(axis=0) calcula la media de cada columna de X.
#X.std(axis=0) calcula la desviación estándar de cada columna de X.
#X_std[X_std==0]=1.0 se utiliza para manejar el caso donde una columna tiene una desviación estándar igual a cero, lo que provocaría una división por cero. Esta línea de código establece esos valores en 1.
#(X-X_mean)/X_std realiza la normalización estándar propiamente dicha, restando la media de cada columna y dividiendo por su desviación estándar.
#Finalmente, la variable X_num se actualiza con los valores normalizados y se devuelve como resultado de la función.

def normalize(X):
    X_mean=X.mean(axis=0)
    X_std=X.std(axis=0)
    X_std[X_std==0]=1.0
    X=(X-X_mean)/X_std
    return X

X_num=normalize(X_num)
X_num.head()
```
