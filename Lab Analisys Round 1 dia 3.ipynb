```python
#Understand the meaning of all columns
customer: number to identify each customer
st: state 
gender: gender of customer
education: level of education
customer lifetime value: measure of the total income a business can expect to bring in from a typical customer for as long as that person or account remains a client (source)
income: income of client
monthly_premium_auto: ampunt of monthly payment the customer makes
number_of_open_complaints: complaints currently open
policy_type: type of policy
vehicle_class: type of vehicle
total_claim_amount: value of a claim
```

```python
#Delete the column education and the number of open complaints from the dataframe.
df1 = df1.drop(['Education','Number of open complaints'],axis=1)
df1
```

```python
#Correct the values in the column customer lifetime value. They are given as a percent, so multiply them by 100 and change dtype to numerical type.
df1['Customer lifetime value']=df1['Customer lifetime value'].replace('%','', regex=True)*100 #.make.str.strip()
df1
```
