```python
#Rearrange the columns in the dataframe as needed
```

```python
file2 = file2[['customer', 'state', 'gender', 'education', 'customer lifetime value',
       'income', 'monthly premium auto', 'number of open complaints',
       'policy type', 'vehicle class', 'total claim amount']]
file2.columns
```

```python
file3 = file3[['customer', 'state', 'gender', 'education', 'customer lifetime value',
       'income', 'monthly premium auto', 'number of open complaints',
       'policy type', 'vehicle class', 'total claim amount']]
file3.columns
```

```python
#Concatenate the three dataframes
```

```python
df=pd.concat([file1,file2,file3],axis=0)
```

```python
df.head(5)
```

```python
def standardize_col(col):
    return col.replace(" ", "_")
```

```python
stdzd_cols = []
for col in df.columns.to_list():
    stdzd_cols.append(standardize_col(col))
stdzd_cols
```

```python
df.columns = stdzd_cols
```
